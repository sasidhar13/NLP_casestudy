```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
###### helper functions. Use them when needed #######
def get_title_from_index(index):
    return df[df.index == index]["Product_name"].values[0]
def get_title_from_ind(index):
    return df[df.index == index]["Product_name"].values[0]
def get_feature_from_ind(index):
    return df[df.index == index]["feature"].values[0]
 
def get_index_from_title(Product_name):
    return df[df.Product_name == Product_name]["index"].values[0]
```

```python
df=pd.read_csv("amazon_phone_dataset.csv")
```

```python
lis=[]
for i in range(9570):
    lis.append(int(i))
lis
```

```python
df['index']=lis
```

```python
features=['feature','prod_des']
for i in features:
    df[i]=df[i].fillna('')
```

```python
df[features]
```

```python
def combine_features(row):
	try:
		return row['feature'] +" "+row['prod_des']
	except:
		print ("Error:", row)	

df["combined_features"] = df.apply(combine_features,axis=1)
df["combined_features"].head()
```

```python
df['combined_features']
```

```python
cv = CountVectorizer()

count_matrix = cv.fit_transform(df["combined_features"])

##Step 5: Compute the Cosine Similarity based on the count_matrix
cosine_sim = cosine_similarity(count_matrix)
```

```python
phone_user_likes = "Samsung Galaxy M40 (Midnight Blue, 6GB RAM, 128GB Storage)"
phone_index=get_index_from_title(phone_user_likes)
phone_index
```

```python
similar_phones =  list(enumerate(cosine_sim[phone_index]))
sorted_similar_phones = sorted(similar_phones,key=lambda x:x[1],reverse=True)
sorted_similar_phones
```
